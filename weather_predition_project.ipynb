## Gradio interface

### Subtask:
Build a Gradio interface to allow users to interact with the trained model or visualize the analysis results.


**Reasoning**:
Define the prediction function and create the Gradio interface.



```python
import gradio as gr
import pandas as pd
import numpy as np

# Define a function that takes input features and returns the predicted temperature
def predict_temperature(apparent_temperature, humidity, wind_speed, wind_bearing, visibility, pressure,
                        loud_cover, precip_type, summary, daily_summary, year, month, day, day_of_week, hour):

    # Create a dictionary from the inputs
    input_data = {
        'Apparent Temperature (C)': apparent_temperature,
        'Humidity': humidity,
        'Wind Speed (km/h)': wind_speed,
        'Wind Bearing (degrees)': wind_bearing,
        'Visibility (km)': visibility,
        'Pressure (millibars)': pressure,
        'Loud Cover': loud_cover,
        'Precip Type': precip_type,
        'Summary': summary,
        'Daily Summary': daily_summary,
        'Year': year,
        'Month': month,
        'Day': day,
        'Day of Week': day_of_week,
        'Hour': hour
    }

    # Convert the dictionary to a pandas DataFrame
    input_df = pd.DataFrame([input_data])

    # One-hot encode categorical features, ensuring all possible columns from training are present
    # Need to recreate the one-hot encoding process from feature engineering
    # Get dummy columns from the original df before splitting
    original_df_for_dummies = pd.read_csv('weatherHistory.csv')
    original_df_for_dummies['Precip Type'].fillna(original_df_for_dummies['Precip Type'].mode()[0], inplace=True)
    original_df_for_dummies = pd.get_dummies(original_df_for_dummies, columns=['Summary', 'Precip Type', 'Daily Summary'], drop_first=True)
    dummy_cols = original_df_for_dummies.columns.tolist()

    # Apply one-hot encoding to the input DataFrame
    input_df = pd.get_dummies(input_df, columns=['Summary', 'Precip Type', 'Daily Summary'], drop_first=True)

    # Align columns with the training data - add missing dummy columns with value 0
    for col in X_train.columns:
        if col not in input_df.columns:
            input_df[col] = 0

    # Ensure the order of columns is the same as X_train
    input_df = input_df[X_train.columns]


    # Predict using the trained model
    prediction = model.predict(input_df)

    return prediction[0]

# Get the unique values for categorical features from the original dataframe
original_df = pd.read_csv('weatherHistory.csv')
summary_choices = original_df['Summary'].unique().tolist()
precip_type_choices = original_df['Precip Type'].fillna(original_df['Precip Type'].mode()[0]).unique().tolist()
daily_summary_choices = original_df['Daily Summary'].unique().tolist()

# Define input components
input_components = [
    gr.Number(label="Apparent Temperature (C)"),
    gr.Number(label="Humidity", minimum=0, maximum=1),
    gr.Number(label="Wind Speed (km/h)", minimum=0),
    gr.Number(label="Wind Bearing (degrees)", minimum=0, maximum=360),
    gr.Number(label="Visibility (km)", minimum=0),
    gr.Number(label="Pressure (millibars)", minimum=0),
    gr.Number(label="Loud Cover", value=0, interactive=False), # Based on data exploration
    gr.Dropdown(label="Precip Type", choices=precip_type_choices),
    gr.Dropdown(label="Summary", choices=summary_choices),
    gr.Dropdown(label="Daily Summary", choices=daily_summary_choices),
    gr.Number(label="Year", minimum=2000, maximum=2025, step=1),
    gr.Number(label="Month", minimum=1, maximum=12, step=1),
    gr.Number(label="Day", minimum=1, maximum=31, step=1),
    gr.Number(label="Day of Week", minimum=0, maximum=6, step=1),
    gr.Number(label="Hour", minimum=0, maximum=23, step=1)
]

# Define output components
output_components = gr.Number(label="Predicted Temperature (C)")

# Create the Gradio interface
iface = gr.Interface(
    fn=predict_temperature,
    inputs=input_components,
    outputs=output_components,
    title="Weather Temperature Predictor",
    description="Predict the temperature based on various weather conditions."
)

# Launch the interface
# iface.launch()
```

Launch the Gradio interface.



```python
iface.launch(share=True)
```
